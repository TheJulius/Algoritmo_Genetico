```python

import random, time

class GABasico():

#METODO INICIALIZADOR
    def __init__(self, tamanho_populacao = 20, geracoes = 1000):

        self.populacao = []
        self.tam_populacao = tamanho_populacao
        self.geracoes = geracoes
        self.soma_avaliacoes = 0

#METODO QUE INICIA A POPULACAO DE CROMOSSOMOS
    def inicializar_populacao(self):

        for i in range(self.tam_populacao):
            self.populacao.append(Cromossomo(10))
        for cromossomo in self.populacao: cromossomo.inicializar()

#METODO QUE VERIFICA A SOMA DAS AVALIACOES
    def avaliar_populacao(self):

        self.soma_avaliacoes = 0
        for cromossomo in self.populacao: #ATRIBUI A SOMA DAS AVALIACOES PARA CADA CROMOSSOMO NA
                                          # NA POPULACAO
            self.soma_avaliacoes += cromossomo.avaliar()

#METODO ROLETA
    def roleta(self):

        self.avaliar_populacao()
        limite = random.random() * self.soma_avaliacoes #LIMITADOR
        i, aux = [0, 0]
        random.shuffle(self.populacao)
        while aux < limite and i < self.tam_populacao:
            aux += self.populacao[i].avaliacao
            i += 1
        i -= 1
        return i

#METODO PARA FAZER A NOVA GERACAO DE CROMOSSOMOS
    def nova_geracao(self):

        nova_populacao = []
        for i in range(self.tam_populacao):
            pai1 = self.populacao[self.roleta()]
            pai2 = self.populacao[self.roleta()]
            filho = pai1.crossover(pai2)
            filho.mutacao(.05) #VALOR DA MUTACAO
            nova_populacao.append(filho)
        return nova_populacao

#EXEC MAIN
    def executar(self):

        print("\n\nInicializando execucao do algorimo genético:...\n\n")
        self.inicializar_populacao()
        self.avaliar_populacao()
        ordenado = sorted(self.populacao, key = lambda x: x.avaliacao, reverse = True)
        for i in range(self.geracoes): #MOSTRA UM TOP 5 MELHORES POSSIVEIS VALORES PARA CADA POPULACAO
            top1 = ordenado[0]
            top2 = ordenado[1]
            top3 = ordenado[2]
            top4 = ordenado[3]
            top5 = ordenado[4]
            print("Ranking dos top 5 cromossomos geração %d:\n%s\n%s\n%s\n%s\n%s\n\n" % (i, top1, top2, top3, top4, top5), end="", flush = True)
            time.sleep(.5)
            
            self.populacao = self.nova_geracao()
            self.avaliar_populacao()
            ordenado = sorted(self.populacao, key = lambda x: x.avaliacao, reverse = True)
        print('top 1 -> %s' % (ordenado[1])) #DIZER QUEM FOI O PRIMEIRO QUE CHEGOU NO MELHOR RESULTADO

#EXECUCAO       
ga = GABasico(tamanho_populacao = 400, geracoes = 10)
ga.executar()
```

```python
import math, random

class Cromossomo():

#INICIALIZADORES CASO NAO SEJA PASSADO NENHUM VALOR
    def __init__(self, tamanho):

        self.tamanho = tamanho
        self.valor = ""
        self.avaliacao = -1

#PEGA VALOR INPUTADO PELO USUARIO
    def set_valor(self, novo_valor):

        self.valor = novo_valor

    def inicializar(self):
        novo_valor = ""

        #CONVERSOR BINARIO
        for i in range(self.tamanho):
            if random.random() > .5: novo_valor += '1'
            else: novo_valor += '0'
        self.set_valor(novo_valor)

#METODO CROSSOVER
    def crossover(self, outro_cromossomo):

        #DEFINE O INDEX DO SPLIT
        split_index = int(random.random() * self.tamanho)
        novo_valor = ""

        #UM LIMITADOR
        if random.random() > .5: #ESTE VALOR CORRESPONDE A CHANCE DE MUTACAO
            novo_valor = self.valor[0:split_index] + outro_cromossomo.valor[split_index:len(outro_cromossomo.valor)]
        else:
            novo_valor = outro_cromossomo.valor[0:split_index] + self.valor[split_index:len(outro_cromossomo.valor)]

        #DEFININDO VALOR DO CROMOSSOMO
        novo_cromossomo = Cromossomo(self.tamanho)
        novo_cromossomo.set_valor(novo_valor)

        return novo_cromossomo

#METODO MUTACAO
    def mutacao(self, chance_mutacao):

        inicio, aux, fim = ['','','']
        for i in range(self.tamanho): # FOR MEDINDO SE O VALOR FOI MENOR QUE A CHANCE DE MUTACAO (0,05%)
            if random.random() < chance_mutacao:
                inicio = self.valor[0:i]
                fim = self.valor[i+1:self.tamanho]
                aux = self.valor[i]
                if aux == '1': aux = '0' #PONTEIROS DE BINARIO
                else: aux = '1'
                self.set_valor(inicio+aux+fim)

#METODO QUE AVALIA SE ESTA DENTRO DA CONDICAO ACEITA
    def avaliar(self):

        x = int(self.valor, 2)

        #FORMULINHA
        self.avaliacao = math.sin(x**3)/(5-math.cos(math.e)-x)
        return self.avaliacao

#METODO QUE MOSTRA OS RESULTADOS
    def __repr__(self):

        return "cromossomo:[%s] avaliacao[%.2f] valor[%d]" % (self.valor, self.avaliacao, int(self.valor, 2))
```
